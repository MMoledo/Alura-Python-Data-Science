```python
import numpy as np
```

```python
dados = np.loadtxt('citrus.csv', delimiter=',',usecols=(np.arange(1,3)),skiprows=1)
dados
```

```python
laranja_diametro = dados[:5000 ,0]
toranja_diametro = dados[5000: ,0]
print(laranja_diametro)
print(toranja_diametro)
```

```python
laranja_peso = dados[:5000 ,1]
toranja_peso = dados[5000: ,1]
print(laranja_peso)
print(toranja_peso)
```

```python
import matplotlib.pyplot as plt

plt.plot(laranja_diametro, laranja_peso, color='orange', label='Laranja')
plt.legend()
plt.xlabel('Di√¢metro')
plt.ylabel('Peso')
plt.show()
```

```python
plt.plot(toranja_diametro, toranja_peso, color='red', label='Toranja')
plt.legend()
plt.xlabel('Di√¢metro')
plt.ylabel('Peso')
plt.show()
```
